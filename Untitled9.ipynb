```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# 1. Check data types and structure
print(df.info())
print(df.head())

# 2. Summarize statistics
print(df.describe())

# 3. Explore data distributions
plt.figure(figsize=(12, 8))
for i, column in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    sns.histplot(df[column], kde=True)
    plt.title(f'Distribution of {column}')
plt.tight_layout()
plt.show()

# 4. Analyze relationships between variables
plt.figure(figsize=(10, 8))
sns.pairplot(df, hue='species')
plt.show()

# Correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm') # Drop the non-numeric column 'species'
plt.title('Correlation Heatmap')
plt.show()
print()

# 5. Handle missing values
print(df.isnull().sum())

# 6. Outlier detection
plt.figure(figsize=(12, 8))
df.boxplot()
plt.title('Box Plots for Outlier Detection')
plt.show()

# Z-score for outlier detection
z_scores = np.abs((df.iloc[:, :4] - df.iloc[:, :4].mean()) / df.iloc[:, :4].std())
print("Potential outliers (Z-score > 3):")
print((z_scores > 3).sum())
```
