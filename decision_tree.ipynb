```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create a CART decision tree classifier
cart_clf = DecisionTreeClassifier(criterion='gini', max_depth=3, random_state=42)

# Train the model
cart_clf.fit(X_train, y_train)

# Make predictions
y_pred = cart_clf.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the parameter grid for CART
param_grid_cart = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'criterion': ['gini']  # Using Gini for CART
}

# Create a CART decision tree classifier
cart_clf = DecisionTreeClassifier(random_state=42)

# Apply Grid Search CV
grid_search_cart = GridSearchCV(cart_clf, param_grid_cart, cv=5, scoring='accuracy')
grid_search_cart.fit(X_train, y_train)

# Get the best model
best_cart = grid_search_cart.best_estimator_

# Make predictions
y_pred_cart = best_cart.predict(X_test)

# Evaluate the model
accuracy_cart = accuracy_score(y_test, y_pred_cart)
print(f'CART Best Parameters: {grid_search_cart.best_params_}')
print(f'CART Accuracy: {accuracy_cart * 100:.2f}%')
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Load the Iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define the parameter grid for ID3
param_grid_id3 = {
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 5],
    'criterion': ['entropy']  # Using Entropy for ID3
}

# Create an ID3 decision tree classifier
id3_clf = DecisionTreeClassifier(random_state=42)

# Apply Grid Search CV
grid_search_id3 = GridSearchCV(id3_clf, param_grid_id3, cv=5, scoring='accuracy')
grid_search_id3.fit(X_train, y_train)

# Get the best model
best_id3 = grid_search_id3.best_estimator_

# Make predictions
y_pred_id3 = best_id3.predict(X_test)

# Evaluate the model
accuracy_id3 = accuracy_score(y_test, y_pred_id3)
print(f'ID3 Best Parameters: {grid_search_id3.best_params_}')
print(f'ID3 Accuracy: {accuracy_id3 * 100:.2f}%')
```
