```python
import pandas as pd
import random
from datetime import datetime, timedelta

# Parameters
num_entries = 10000
categories = ['Food', 'Entertainment', 'Utilities', 'Transportation', 'Health', 'Clothing', 'Miscellaneous']
payment_methods = ['Credit Card', 'Cash', 'Debit Card', 'Mobile Payment']

# Generate data
data = {
    "ID": range(1, num_entries + 1),
    "Date": [(datetime.now() - timedelta(days=random.randint(0, 365))).strftime('%Y-%m-%d') for _ in range(num_entries)],
    "Category": [random.choice(categories) for _ in range(num_entries)],
    "Amount": [round(random.uniform(5.0, 500.0), 2) for _ in range(num_entries)],
    "Payment Method": [random.choice(payment_methods) for _ in range(num_entries)],
    "Location": [random.choice(['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']) for _ in range(num_entries)]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv('personal_spending_dataset.csv')
```
